CHECKING THE MODEL`S ACCURACY,CREATING THE CONFUSION MATRIX AND PLOTTING CONFUSION MATRIX`S HEATMAP

```python
from sklearn.metrics import accuracy_score,confusion_matrix
y_pred=tree_model_fitted.predict(x_test)
cm=confusion_matrix(y_test,y_pred)
accuracy=accuracy_score(y_test,y_pred)*100
print("Accuracy:",accuracy)
print(cm)





#HEATMAP
plt.figure(figsize=(6,4))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", cbar=False)
plt.xlabel("predicted Labels")
plt.ylabel("actual Lables")
```

CHECKING THE MODEL`S PRECISION, RECALL, F1-SCORE AND CLASSIFICATION REPORT

```python
from sklearn.metrics import precision_score,f1_score,recall_score,classification_report
precision=precision_score(y_test,y_pred, average='weighted')*100
f1=f1_score(y_test,y_pred, average='weighted')*100
recall=recall_score(y_test,y_pred, average='weighted')*100

print("Precision:", precision)
print("fi:", f1)
print("recall:", recall)

print("\nClassification Report:\n", classification_report(y_test,y_pred))
```

PLOTTING THE DECISION TREE
